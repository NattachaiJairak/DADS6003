```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Read the CSV file from the URL
url = "https://raw.githubusercontent.com/ekaratnida/Applied-machine-learning/master/Week14-kmeans/Mall_Customers.csv"
data = pd.read_csv(url)

# Reduce some unnecessary columns
data_reduce = data.drop(['CustomerID', 'Gender'], axis=1)

# Normalize the data
data_norm = (data_reduce - data_reduce.mean()) / data_reduce.std()

# Fit k-means clustering with the optimal number of clusters (let's say 5 for example)
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
data_norm['Cluster'] = kmeans.fit_predict(data_norm)

# Get centroids
centroids = kmeans.cluster_centers_

# Define colors for centroids
centroid_colors = ['blue', 'green', 'red', 'orange', 'purple']

# Plot scatter plot for each cluster and centroids
plt.figure(figsize=(10, 6))
for cluster_label in range(5):
    cluster_data = data_norm[data_norm['Cluster'] == cluster_label]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {cluster_label}')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color=centroid_colors, label='Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clusters with Centroids')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Read the CSV file from the URL
url = "https://raw.githubusercontent.com/ekaratnida/Applied-machine-learning/master/Week14-kmeans/Mall_Customers.csv"
data = pd.read_csv(url)

# Reduce some unnecessary columns
data_reduce = data.drop(['CustomerID', 'Gender'], axis=1)

# Normalize the data
data_norm = (data_reduce - data_reduce.mean()) / data_reduce.std()

# Fit k-means clustering with the optimal number of clusters (let's say 5 for example)
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
data_norm['Cluster'] = kmeans.fit_predict(data_norm)

# Get centroids
centroids = kmeans.cluster_centers_

# Define colors for centroids
centroid_colors = ['blue', 'green', 'red', 'orange', 'purple']

# Plot scatter plot for each cluster and centroids
plt.figure(figsize=(10, 6))
for cluster_label in range(5):
    cluster_data = data_norm[data_norm['Cluster'] == cluster_label]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {cluster_label}')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color=centroid_colors, label='Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clusters with Centroids')
plt.legend()
plt.grid(True)
plt.show()

# Plot bar chart for sum of each feature within each cluster
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for feature in features:
    plt.figure(figsize=(8, 5))
    for cluster_label in range(5):
        cluster_data = data[data_norm['Cluster'] == cluster_label]
        plt.bar(cluster_label, cluster_data[feature].sum(), color=centroid_colors[cluster_label], label=f'Cluster {cluster_label}')
    plt.xlabel('Cluster')
    plt.ylabel('Sum')
    plt.title(f'Sum of {feature} for Each Cluster')
    plt.xticks(range(5), range(5))
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Read the CSV file from the URL
url = "https://raw.githubusercontent.com/ekaratnida/Applied-machine-learning/master/Week14-kmeans/Mall_Customers.csv"
data = pd.read_csv(url)

# Perform k-means clustering
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
data['Cluster'] = kmeans.fit_predict(data[['Annual Income (k$)', 'Spending Score (1-100)']])

# Get centroids
centroids = kmeans.cluster_centers_

# Define colors for clusters
cluster_colors = ['blue', 'green', 'red', 'orange', 'purple']

# Plot scatter plot for Annual Income vs Spending Score with clusters and centroids
plt.figure(figsize=(10, 6))
for cluster_label in range(5):
    cluster_data = data[data['Cluster'] == cluster_label]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], color=cluster_colors[cluster_label], label=f'Cluster {cluster_label}')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='black', label='Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clusters with Centroids')
plt.legend()
plt.grid(True)
plt.show()
```
