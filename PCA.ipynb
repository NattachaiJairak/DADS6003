```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_regression(n_samples=100, n_features=20, random_state=1234)

# PCA
pca = PCA(n_components=0.95)
pca.fit(X)

# Explained variance ratio
#print("pca.explained_variance_ratio_", pca.explained_variance_ratio_)

# Cumulative explained variance
cs = np.cumsum(pca.explained_variance_ratio_)
#print("Cumulative explained variance:", cs)

# Explained variance
#print("pca.explained_variance_", pca.explained_variance_)

# Principal components
#print("pca.components_", pca.components_)

# Plotting cumulative explained variance
plt.figure()
plt.bar(range(len(cs)), cs)
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by Principal Components')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.decomposition import PCA

# Generate synthetic data
X, y = make_regression(n_samples=100, n_features=20, random_state=1234)

# Create a PCA model that retains 95% of the variance
model = PCA(n_components=0.95)
model.fit(X)

# Transform the data to reduced dimensionality
X_pca = model.transform(X)

# Print explained variance ratio
#print("Explained variance ratio:", model.explained_variance_ratio_)

# Calculate cumulative variance explained
cumulative_variance = np.cumsum(model.explained_variance_ratio_)
#print("Cumulative variance explained:", cumulative_variance)

# Print explained variance (individual components)
#print("Explained variance (individual components):", model.explained_variance_)

# Plot cumulative variance explained
plt.figure(figsize=(8, 6))
plt.bar(range(len(cumulative_variance)), cumulative_variance)
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Variance Explained")
plt.title("Cumulative Variance Explained by Principal Components")
plt.show()
```
